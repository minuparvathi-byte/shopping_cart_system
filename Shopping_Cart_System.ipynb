# Case Study – Shopping Cart System
Instructions:

Create a list named cart with initial items: Milk, Bread, Eggs, Apple

Perform list operations:

• Add multiple items using extend()

• Insert item at index 2 using insert()

• Remove item Apple using remove()

• Remove last item using pop()

Use list methods:

• index() to find position of Eggs

• count() to count occurrences of Milk

Apply indexing:

• Display first item

• Display last item

Apply slicing:

• Display first 3 items from the list

Find total number of items using len()

Display all outputs using print()

```python
Shopping_cart = ["Milk","Bread","Eggs","Apple"]

 #extend
Shopping_cart.extend(["Oil","Jam"])
print(f"Shopping_list:{Shopping_cart}")


#insert()

Shopping_cart.insert(2,"Batter")

print(f"Shopping_list:{Shopping_cart}")

#remove()
Shopping_cart.remove("Apple")
print(f"Shopping_list:{Shopping_cart}")


#pop()

Shopping_cart.pop()
print(f"Shopping_list:{Shopping_cart}")


#list methods

print("Index of Eggs:",Shopping_cart.index("Eggs"))

#count
print("count of Milk:",Shopping_cart.count("Milk"))

#indexing and slicing
print("First item:",Shopping_cart[0])

print("Last item:",Shopping_cart[-1])

print("First three item:",Shopping_cart[0:3])

#length
print("Total no of items:",len(Shopping_cart))


print(f"Shopping_list:{Shopping_cart}")


```
